```python
#for converting dictionaries inside the string to proper dictionaries
import ast
```

```python
#for taking the names from the given dictnory and converting it into list
def convert_genres(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies_df['genres']=movies_df['genres'].apply(convert_genres)
```

```python
#for taking the names of 1st 3 persons from the given dictnory and converting it into list
def convert_cast(obj):
    L=[]
    ct=0
    for i in ast.literal_eval(obj):
        if ct!=3:
            L.append(i['name'])
            ct+=1

    return L
```

```python
movies_df['cast']=movies_df['cast'].apply(convert_cast)
```

```python
ast.literal_eval(movies_df['crew'][0])
```

```python
#for taking the names from the given dictnory and converting it into list
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
#for taking the names of 1st 3 persons from the given dictnory and converting it into list
def convert_keywords(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])

    return L
```

stemming is the method of getting the root word from the word
eg: love,loved,loving it's root word is love 

```python
#stem function
def stem(text):
    lst=[]
    for i in text.split():
        lst.append(ps().stem(i))
    return " ".join(lst)
```

```python
new_movies_df['tags']=new_movies_df['tags'].apply(stem)
```

```python
#function to resolve memmory error while doing cosine similarity
#no idea how it works, got it from chatgpt
def compute_cosine_similarity_in_chunks(X, chunk_size=1000):
    num_chunks = X.shape[0] // chunk_size + (X.shape[0] % chunk_size != 0)
    num_dims = X.shape[1:]
    similarities = np.zeros((X.shape[0], X.shape[0], *num_dims))
    for i in range(num_chunks):
        start = i * chunk_size
        end = (i + 1) * chunk_size
        chunk = X[start:end, ...]
        chunk_similarities = cosine_similarity(chunk.reshape(chunk.shape[0], -1), X.reshape(X.shape[0], -1))
        similarities[start:end, ...] = chunk_similarities.reshape(*chunk_similarities.shape[:2], *num_dims)
    return similarities
```

```python
similarity_matrix=compute_cosine_similarity_in_chunks(movies_vector)
```
